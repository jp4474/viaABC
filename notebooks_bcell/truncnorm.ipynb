```python
from scipy.stats import truncnorm
import numpy as np
```

```python
# Parameters
lower = np.array([0, 0, 0, 0, 0, 0])
upper = np.array([1, 0.1, 1, 1, 1, 1])
mean = np.array([0.01, 0.01, 0.01, 0.01, 0.01, 0.01])
std = np.array([0.05, 0.05, 0.05, 0.05, 0.05, 0.05])
size = 1000

# Standardize the bounds
a, b = (lower - mean) / std, (upper - mean) / std

# Generate samples
x = truncnorm(a, b, loc=mean, scale=std).rvs(1)
```

```python
from matplotlib import pyplot as plt
```

```python

fig, ax = plt.subplots(1, 1)  # Correct function and use of subplots

ax.hist(x, bins = 100)  # Use ax directly, not ax[0]

plt.show()  # Optional, but necessary to display the plot in some environments
```
