```python
from scipy.integrate import solve_ivp
import os

os.chdir('/home/jp4474/latent-abc-smc')
from systems import LotkaVolterra
```

```python
lotka_abc = LotkaVolterra()
```

```python
trajectories = np.zeros((1000, 2, 1500))
for i in range(1000):
    # choose final particles
    param_set = df.iloc[i][['theta.1', 'theta.2']].values
    solution = solve_ivp(lotka_abc.ode_system, [lotka_abc.t0, lotka_abc.tmax], y0=lotka_abc.state0, t_eval=np.arange(0, 15, 0.01), args=(param_set,))
    trajectories[i] = solution.y
trajectories.shape
```

```python
x = np.arange(0, 15, 0.01)
mean_trajectories = trajectories.mean(axis=0)
mean_trajectories = np.percentile(trajectories, 50, axis = 0)
```

```python
confidence_interval = 0.95
lower_bound = np.percentile(trajectories, 2.5, axis=0)
upper_bound = np.percentile(trajectories, 97.5, axis=0)
```

```python
lower_bound
```

```python
observational_data_trajectory = np.load("data/Lotka/lotka_data.npz")["obs_data"]
ground_truth_trajectory = solve_ivp(lotka_abc.ode_system, [lotka_abc.t0, lotka_abc.tmax], y0=lotka_abc.state0, t_eval=np.arange(0, 15, 0.01), args=([1.,1.],)).y
```

```python
lower_bound[1]
```

```python
upper_bound[1]
```

```python
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot prey
ax[0].plot(x, mean_trajectories[0], label="Median", linewidth=2)
ax[0].fill_between(x, lower_bound[0], upper_bound[0], alpha=0.3, label="95% Credible Interval")
ax[0].scatter(lotka_abc.time_space, observational_data_trajectory[:,0], label="Observations", s=30, zorder=5, c='C0')
ax[0].plot(x, ground_truth_trajectory[0], label="Ground Truth", color='black', linestyle='--', linewidth=2)

# Plot predator
ax[1].plot(x, mean_trajectories[1], label="Median", color="orange", linewidth=2)
ax[1].fill_between(x, lower_bound[1], upper_bound[1], alpha=0.3, color="orange", label="95% Credible Interval")
ax[1].scatter(lotka_abc.time_space, observational_data_trajectory[:,1], label="Observations", s=30, color="orange", zorder=5)
ax[1].plot(x, ground_truth_trajectory[1], label="Ground Truth", color='black', linestyle='--', linewidth=2)


# Set labels and titles
ax[0].set_ylabel("Population")
ax[0].set_xlabel("Time")
ax[1].set_xlabel("Time")
ax[0].set_title("Prey")
ax[1].set_title("Predator")

# Add grid for better visibility
ax[0].grid(True, alpha=0.3)
ax[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(f"/home/jp4474/latent-abc-smc/notebooks_lotka/figures/stan_fit.pdf", dpi=300)
```
