```python
for i, generation in enumerate(generations):
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))

    weights = generation['weights']

    # Plot weighted posterior histogram for Beta
    ax[0].hist(generation['particles'][:, 0], bins=30, weights=weights, density=False,
               alpha=0.5, label='Posterior')
    ax[0].set_title(f'Posterior of Beta at Generation {i}')
    ax[0].set_xlabel('Value')
    ax[0].set_ylabel('Density')
    ax[0].legend()
    ax[0].axvline(1.0, color='red', linestyle='dashed', linewidth=1, label='True Beta')

    # Plot weighted posterior histogram for Tau_I
    ax[1].hist(generation['particles'][:, 1], bins=30, weights=weights, density=False, alpha=0.5, label='Posterior')
    ax[1].set_title(f'Posterior of Tau_I at Generation {i}')
    ax[1].set_xlabel('Value')
    ax[1].set_ylabel('Density')
    ax[1].legend()
    ax[1].axvline(1.0, color='red', linestyle='dashed', linewidth=1, label='True Tau_I')

    plt.tight_layout()
    # plt.show()
    plt.savefig(f"{path}/generation_{i}.png")
    plt.close(fig)
```
